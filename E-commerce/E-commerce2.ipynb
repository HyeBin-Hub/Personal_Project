```python
import pandas as pd
import matplotlib.pyplot as plt

clothing=pd.read_csv("./data/Womens Clothing E-Commerce Reviews.csv")
clothing
```

```python
# 25살을 기준으로 Clothing ID의 개수를 pivot_table로 만듬
pivot=pd.pivot_table(clothing,index="Clothing ID",values="Title",columns=["Age"],aggfunc="count")
pivot=pivot.fillna(0)

top10=pivot.sort_values(25,ascending=False).head(10)
top10
```

```python
# 25살을 기준으로 Clothing ID의 개수 top10 시각화 
fig,axes=plt.subplots(1,figsize=(13,5))

for i in top10.iloc:
    axes.plot(i.index,i.to_numpy(),label=i.name)
axes.set_xticklabels(axes.get_xticks(),size=13)
axes.set_yticklabels(axes.get_yticks(),size=13)
axes.legend(loc=1,fontsize=10)
axes.set_title("Visualize top 10 number of Clothing ID as of age 25",size=17,pad=15)
```

```python
# 18세에서 50세 사이에서 25살을 기준으로 해당 Clothing ID의 개수 top10 시각화 
fig,axes=plt.subplots(1,figsize=(10,5))
top10=pivot[range(18,50)].sort_values(25,ascending=False).head(10)

for i in top10.iloc:
    axes.plot(i.index,i.to_numpy(),label=i.name)
axes.set_xticklabels(axes.get_xticks(),size=13 )
axes.set_yticklabels(axes.get_yticks(),size=13 )
axes.legend(loc=2,fontsize=10)
axes.set_title("Visualize top10 number of Clothing ID by age between 18 and 50 and 25",size=17,pad=15)
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
import nltk
import re
# nltk
#  : 말뭉치(corpus), 토큰 생성, 형태소 분석, 품사 태깅등의 자연어처리 및 문서 분석 패키지

# 데이터 전처리 함수 생성
def text_pro(texts):
    # 영어 불용어 리스트를 불러온다
    stop_words=set(nltk.corpus.stopwords.words("english"))
    result=[]
    for text in texts:
        text=re.sub("[-=+,#/\?:';^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]","",text)
        words=text.split(" ")
        text=" ".join([w for w in words if w not in stop_words])
        result.append(text)
    return result

# 데이터 전처리
clothing["Review Text"]=clothing["Review Text"].fillna("")
clothing["Review Text"]=clothing["Review Text"].astype("string")

review=text_pro(clothing["Review Text"])
clothing["Review Text"]=review

clothing['Class Name'] = clothing['Class Name'].fillna('None')
clothing['Division Name'] = clothing['Division Name'].fillna('None')
clothing['Age'] = clothing['Age'].fillna(30) 
```

```python
train = clothing[['Age','Division Name','Class Name',"Review Text",'Rating']].to_numpy()
X_train, X_test, y_train, y_test = train_test_split(train[:,:-1], train[:,-1], test_size=0.1)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
adapt_age = np.array(range(100), dtype=np.float32)
adapt_age = np.reshape(adapt_age,(-1,1))
cn_nq = clothing['Class Name'].unique()
dn_nq = clothing['Division Name'].unique()

print(adapt_age.shape)
print(cn_nq)
print(dn_nq)
```

- GRU(Gated Reccurent Unit)

<img src="./image/GRU.png" height="400px" width="400px" >

1. Reset Gate\
    : 이전 hidden state의 값을 얼마나 활용할 것인지에 대한 제어
2. Update Gate\
    : 과거와 현재의 정보를 각각 얼마나 반영할지에 대한 비율을 구하는 것
    - LSTM의 input,forget gate 모두를 제어한다 
- cell state와 hidden state가 합쳐져 하나의 hidden state로 표현된다

```python
import tensorflow as tf

# text data 층 추가
# 입력으로 4개가 들어간다
age = tf.keras.Input(shape=(1,), name='Age')
cn = tf.keras.Input(shape=(1,), dtype='string', name='ClassName')
dn = tf.keras.Input(shape=(1,), dtype='string', name='DivisionName')
text=tf.keras.Input(shape=(1,),dtype="string",name="Text")

# age 
norm = tf.keras.layers.experimental.preprocessing.Normalization()
# adapt()은 임의의 배열을 만들어 그 배열을 normalization하여 정규화하고자 하는 데이터에 적용
norm.adapt(adapt_age)
age_norm = norm(age)

# text
# TextVectorization는 텍스트 벡터화 layer이다
# output_mode 인자 : 어떻게 벡터화할지 정한다("int", "binary", "count"또는 "tf-idf")
#                - int : 분할 문자열 토큰 당 하나의 정수 인덱스를 출력한다
# output_sequence_length 인자 : int 모드에서만 사용할 수 있으며, 출력의 길이를 의미한다
vec_layer=tf.keras.layers.experimental.preprocessing.TextVectorization(
    output_mode="int",
    output_sequence_length=30)

vec_layer.adapt(review)
text_vec=vec_layer(text)
vocab_size=len(vec_layer.get_vocabulary())
text_em=tf.keras.layers.Embedding(vocab_size,100)(text_vec) # batch,30,100

# GRU는 LSTM보다 구조가 단순하지만, 성능면에서는 비슷하다고 한다
gru1_layer=tf.keras.layers.GRU(128,return_sequences=True)
gru2_layer=tf.keras.layers.GRU(64,)

gru1=tf.keras.layers.Bidirectional(gru1_layer)(text_em) # batch,30,128*2
gru2=tf.keras.layers.Bidirectional(gru2_layer)(gru1) # batch, 64*2

# StringLookup :미리 만들어놓은 cn_nq를 이용해 입력 데이터를 index로 표현한다
cn_look = tf.keras.layers.experimental.preprocessing.StringLookup(vocabulary=cn_nq, mask_token=None)(cn)
cn_em = tf.keras.layers.Embedding(len(cn_nq) + 1, 20)(cn_look) # (batch,1,10)

# StringLookup :미리 만들어놓은 dn_nq를 이용해 입력 데이터를 index로 표현한다
dn_look = tf.keras.layers.experimental.preprocessing.StringLookup(vocabulary=dn_nq, mask_token=None)(dn)
dn_em = tf.keras.layers.Embedding(len(dn_nq) + 1, 5)(dn_look) # (batch,1, 5)

concat_em = tf.keras.layers.Concatenate()([cn_em, dn_em]) # (batch,1, 15)

flatten = tf.keras.layers.Flatten()(concat_em)

dc_dense = tf.keras.layers.Dense(16, activation="relu")(flatten) # batch,16
age_dense = tf.keras.layers.Dense(8, activation="relu")(age_norm) # batch, 8
text_dense=tf.keras.layers.Dense(16,activation="relu")(gru2) # batch, 16

total = tf.keras.layers.Concatenate()([dc_dense, age_dense,text_dense]) # batch,16+8+16

d1 = tf.keras.layers.Dense(16, activation="relu")(total)
d2 = tf.keras.layers.Dense(1)(d1)


model = tf.keras.Model(inputs = [age, cn, dn,text], outputs = d2 )
model.summary()
```

```python
# 모델 학습과정 설정
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
              loss= tf.keras.losses.MeanSquaredError())
```

```python
age_train = X_train[:,0:1].astype(np.float32)
dn_train = X_train[:,1:2]
cn_train = X_train[:,2:3]
text_train=X_train[:,3:4]
y_train = y_train.astype(np.float32)
print(age_train.shape)
print(dn_train.shape)
print(cn_train.shape)
```

```python
from tensorflow.keras.callbacks import EarlyStopping

# 저장할 위치 지정
filepath="./E-Commerce"
# ModelCheckpoint : 모델이 학습하면서 정의한 조건을 만족했을 때 Model의 weight 값을 중간 저장
# 인자
#  - filepath : 모델을 저장할 경로
#  - monitor : 모델을 저장할때 기준이 되는 값을 지정
#  - save_best_only = True : monitor에서 지정한 값이 가장 좋은 값일때 저장
#    save_best_only = False : 매 에폭마다 모든 값 저장
#  - save_weights_only = True : weights만 저장
#    save_weights_only = False : 레이어 및 weights 둘다 저장
#  -  mode : val_acc는 정확도가 높을수록 좋기 때문에 mode를 max로 입력하고
#            val_loss는 정확도가 낮을수록 좋기 때문에 min으로 입력한다
#            max,min을 자동으로 입력하고 싶을땐 auto사용
mc=tf.keras.callbacks.ModelCheckpoint(filepath,monitor="val_loss",
                                      save_best_only=True,
                                      save_weights_only=True,
                                      mode="auto")

# EarlyStopping : 모니터링 된 지표가 개선을 멈췄을 때 훈련을 중지
es=tf.keras.callbacks.EarlyStopping(monitor="val_loss", mode="auto",patience=3)

# 모델 학습
hist=model.fit({'Age':age_train, 'ClassName': cn_train, 'DivisionName':dn_train,"Text":text_train},
          y_train,
          batch_size=100,
          epochs=20,
          callbacks=[es,mc],
          validation_split=0.2)
```

```python
import matplotlib.pyplot as plt

fig,loss_ax=plt.subplots()
acc_ax=loss_ax.twinx()

loss_ax.plot(hist.history["loss"],"y",label="train loss")
loss_ax.plot(hist.history["val_loss"],"r",label="val loss")
loss_ax.set_xlabel("epoch")
loss_ax.set_ylabel("loss")
loss_ax.legend(loc="upper right")

# acc_ax.plot(hist.history["accuracy"],"b",label="train acc")
# acc_ax.plot(hist.history["val_accuracy"],"g",label="val acc")
# acc_ax.set_ylabel("accuracy")
# acc_ax.legend(loc="upper left")

plt.show()
```

```python
age_test = X_test[:,0:1].astype(np.float32)
dn_test = X_test[:,1:2]
cn_test = X_test[:,2:3]
text_test=X_test[:,3:4]

y_test = y_test.astype(np.float32) # (2349,)

# ModelCheckpoint에서 저장한 모델을 가지고 다시 평가한다
model.load_weights(filepath)

# 모델 예측
result=model.predict({'Age':age_test, 'ClassName': cn_test, 'DivisionName':dn_test,"Text":text_test})
print(result) # (2349,1)
# 배열에서 길이가 1차원인 것을 제거한다
predict=np.squeeze(result) #  (2349,)
# 각각의 배열들을 새로운 축으로 합친다(배열들은 같은 shape을 가져야한다)
predict_set=np.stack((y_test,predict),axis=1) # (2349,2)
```

```python
predict_df=pd.DataFrame(predict_set,columns=["y","predict"])
predict_df.head()
```
